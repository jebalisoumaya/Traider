```python
# =============================
# 🔹 1. Imports & Setup
# =============================
%pip install numpy==1.26.4 tensorflow==2.15.0 scikit-learn==1.4.2 yfinance==0.2.40 --upgrade --force-reinstall

import yfinance as yf
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import subprocess, shlex

from tensorflow.keras.models import Sequential
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator

# Vérifier l'utilisation du GPU
print("TensorFlow version:", tf.__version__)
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    print(f"{len(gpus)} GPU(s) détecté(s) :")
    for i, gpu in enumerate(gpus):
        print(f" - GPU {i}: {gpu}")
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        logical_gpus = tf.config.list_logical_devices('GPU')
        print("Logical GPUs:", len(logical_gpus))
    except Exception as e:
        print("Impossible d'activer memory_growth:", e)
else:
    print("Aucun GPU détecté par TensorFlow.")

try:
    out = subprocess.check_output(shlex.split("nvidia-smi -L"), stderr=subprocess.STDOUT, text=True)
    print("nvidia-smi:\n", out)
except Exception as e:
    print("nvidia-smi non disponible ou erreur:", e)

# -----------------------------
# Paramètres
# -----------------------------
symbol = "BTC-USD"
interval = "1h"   # Intervalle horaire
period = "60d"    # 60 jours d’historique

# -----------------------------
# Téléchargement des données
# -----------------------------
df = yf.download(tickers=symbol, interval=interval, period=period)

if df.empty:
    raise ValueError(f"Aucune donnée téléchargée pour {symbol}")

print(f"{df.shape[0]} lignes téléchargées")
df.head()

# =============================
# 🔹 2. Préparation des données (avec OHLCV)
# =============================

features_cols = ['Open', 'High', 'Low', 'Close', 'Volume']
data = df[features_cols].astype('float32')

# Normalisation
scaler = MinMaxScaler(feature_range=(0, 1))
data_normalized = scaler.fit_transform(data)

print(f"✅ Données normalisées avec {len(features_cols)} features : {features_cols}")
print(f"Shape normalisé : {data_normalized.shape}")

# =============================
# 🔹 3. Génération des séquences
# =============================

look_back = 72
split_idx = int(len(data_normalized) * 0.9)

train_generator = TimeseriesGenerator(
    data_normalized, data_normalized[:, 3],  # On prédit le 'Close'
    length=look_back, batch_size=32,
    start_index=0, end_index=split_idx - 1
)

val_generator = TimeseriesGenerator(
    data_normalized, data_normalized[:, 3],
    length=look_back, batch_size=32,
    start_index=split_idx - look_back
)

print(f"Générateurs créés avec {data_normalized.shape[1]} features et look_back={look_back}")
print(f"Nombre de batches: {len(train_generator)}")

# =============================
# 🔹 4. Modèle LSTM
# =============================

model = Sequential([
    LSTM(50, input_shape=(look_back, len(features_cols))),  # 5 features
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()

# =============================
# 🔹 5. Entraînement du LSTM
# =============================

history = model.fit(train_generator, validation_data=val_generator, epochs=50, verbose=1)

# =============================
# 🔹 6. Prédictions et visualisation (LSTM)
# =============================

test_generator = TimeseriesGenerator(
    data_normalized, data_normalized[:, 3],
    length=look_back, batch_size=1
)

predictions_normalized = model.predict(test_generator)

# 🧩 Correction ici : inverse_transform spécifique à la colonne Close
temp = np.zeros((len(predictions_normalized), len(features_cols)))
temp[:, 3] = predictions_normalized[:, 0]
predictions = scaler.inverse_transform(temp)[:, 3].reshape(-1, 1)

vraies_valeurs = data['Close'].values[look_back:].reshape(-1, 1)

# ---- Graphiques ----
plt.figure(figsize=(8,4))
plt.plot(history.history.get('loss', []), label='train_loss')
if 'val_loss' in history.history:
    plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Courbe de loss LSTM')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(vraies_valeurs, label='Vraies Valeurs', alpha=0.8)
plt.plot(predictions, label='Prédictions LSTM', alpha=0.7)
plt.title(f"BTC-USD — Vraies vs Prédictions (LSTM) — look_back={look_back}")
plt.xlabel('Temps')
plt.ylabel('Prix (Close)')
plt.legend()
plt.tight_layout()
plt.show()

mse = np.mean((vraies_valeurs - predictions)**2)
mae = np.mean(np.abs(vraies_valeurs - predictions))
print(f"LSTM → MSE: {mse:.2f} | MAE: {mae:.2f}")

# =============================
# 🔹 7. Modèle TFT (Temporal Fusion Transformer simplifié)
# =============================

from tensorflow.keras import layers, Model, Input

print("\n🚀 Entraînement d’un petit modèle TFT simplifié...\n")

# ---- Données pour TFT ----
X_train_batches, y_train_batches = zip(*train_generator)
X_val_batches,   y_val_batches   = zip(*val_generator)
X_train = np.concatenate(X_train_batches, axis=0)
y_train = np.concatenate(y_train_batches, axis=0)
X_val   = np.concatenate(X_val_batches,   axis=0)
y_val   = np.concatenate(y_val_batches,   axis=0)

print(f"Forme X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"Forme X_val:   {X_val.shape},   y_val:   {y_val.shape}")

# ---- Architecture TFT simplifiée ----
inp = Input(shape=(look_back, 1))
x = layers.LSTM(32, return_sequences=True)(inp)
x = layers.MultiHeadAttention(num_heads=4, key_dim=8)(x, x)
x = layers.LayerNormalization()(x)
x = layers.Dense(64, activation='relu')(x)
x = layers.Dropout(0.2)(x)
gate = layers.Dense(64, activation='sigmoid')(x)
x = layers.Multiply()([x, gate])
x = layers.GlobalAveragePooling1D()(x)
out = layers.Dense(1)(x)

tft_model = Model(inp, out, name="TFT_Simplifie")
tft_model.compile(optimizer='adam', loss='mean_squared_error')
tft_model.summary()

# ---- Entraînement ----
history_tft = tft_model.fit(
    X_train[:, :, 3].reshape((-1, look_back, 1)),  # on donne seulement la série Close
    y_train,
    validation_data=(X_val[:, :, 3].reshape((-1, look_back, 1)), y_val),
    epochs=50,
    batch_size=32,
    verbose=1
)

# ---- Prédictions TFT ----
X_all_batches, y_all_batches = zip(*test_generator)
X_all = np.concatenate(X_all_batches, axis=0)
y_all = np.concatenate(y_all_batches, axis=0)
tft_pred_norm = tft_model.predict(X_all[:, :, 3].reshape((-1, look_back, 1)))

# 🧩 Correction ici aussi
temp2 = np.zeros((len(tft_pred_norm), len(features_cols)))
temp2[:, 3] = tft_pred_norm[:, 0]
tft_pred = scaler.inverse_transform(temp2)[:, 3].reshape(-1, 1)

# ---- Visualisation comparative ----
plt.figure(figsize=(12, 6))
plt.plot(vraies_valeurs, label='Vraies Valeurs', alpha=0.8)
plt.plot(predictions, label='LSTM', alpha=0.7)
plt.plot(tft_pred, label='TFT (simplifié)', alpha=0.7)
plt.title(f"Comparaison LSTM vs TFT sur BTC-USD ({interval})")
plt.xlabel('Temps')
plt.ylabel('Prix (Close)')
plt.legend()
plt.tight_layout()
plt.show()

mse_tft = np.mean((vraies_valeurs - tft_pred)**2)
mae_tft = np.mean(np.abs(vraies_valeurs - tft_pred))
print(f"TFT → MSE: {mse_tft:.2f} | MAE: {mae_tft:.2f}")
print("\n✅ TFT entraîné et comparé avec succès !")

# =============================
# 🔹 8. Préparation des signaux pour l'agent DARQN
# =============================

print("\n💾 Préparation des signaux pour le futur agent DARQN...\n")

min_len = min(len(vraies_valeurs), len(predictions), len(tft_pred))
vraies_valeurs = vraies_valeurs[-min_len:]
predictions = predictions[-min_len:]
tft_pred = tft_pred[-min_len:]

# Features : on prend les 5 colonnes OHLCV normalisées (dernière période alignée)
features_test = data_normalized[-min_len-look_back:-look_back]

# Signaux : les prédictions des deux modèles
signals_test = np.concatenate([
    predictions[-min_len:],     
    tft_pred[-min_len:]
], axis=1)

# Prix réels
prices_test = vraies_valeurs[-min_len:].reshape(-1, 1)

print(f"✅ features_test.shape = {features_test.shape}")
print(f"✅ signals_test.shape  = {signals_test.shape}")
print(f"✅ prices_test.shape   = {prices_test.shape}")

# Sauvegarde du fichier .npz
np.savez(
    "btc_dataset_for_agent.npz",
    features=features_test,
    signals=signals_test,
    prices=prices_test
)

print("\n🎯 Fichier 'btc_dataset_for_agent.npz' généré avec succès !")
print("Contient : features, signals, prices — prêt pour le Notebook 2 (DARQN).")

# Vérification
data_check = np.load("btc_dataset_for_agent.npz")
print("\nClés disponibles :", list(data_check.keys()))
print("Exemple de ligne feature:", data_check['features'][0])
print("Exemple de ligne signal :", data_check['signals'][0])
print("Exemple de prix         :", data_check['prices'][0])
```
